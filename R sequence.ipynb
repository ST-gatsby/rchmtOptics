```python
def phi(d): 
    x=2.0000 
    for i in range(10): 
        x = pow(1+x,1/(d+1)) 
    return x

def Rseq(d,n):
    g = phi(d)
    a1 = 1.0/g
    a2 = 1.0/(g*g)
    X, Y = [],[]
    
    for i in range(1,n):
        X.append((0.5+a1*i) %1)
        Y.append((0.5+a2*i) %1)
        
    return X,Y
```

```python
import matplotlib.pyplot as plt

# Number of dimensions. 
d=2 

# number of required points 
n=10000

X,Y = Rseq(d,n)
plt.scatter(X,Y, s=0.5, c='black')
```

```python
from math import floor
def haltonSeq(n,b):
    X = []
    
    r, f = 0, 1
    while n>0:
        f = f/b
        r=r+f*(n%b)
        n=floor(n/b)
        X.append(r)
        
    return X
```

```python
# Nombre de point requis
n=1000

# Base de nombres premiers
BX=2
BY=3

X = haltonSeq(n,BX)
Y = haltonSeq(n,BY)
print(X[:10])

#plt.scatter(X,Y, s=0.5, c='black')
```
